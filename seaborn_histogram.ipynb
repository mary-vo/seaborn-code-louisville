## Intro Visuals

### Basics

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(penguins.bill_length_mm)
```

```python
sns.histplot(x= 'bill_length_mm', data=penguins)
```

```python
# To make this a horizontal chart, change 'x' to y
sns.histplot(y= 'bill_length_mm', data=penguins)
```

```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             kde=True)
```

```python
# By default, seaborn will decide ho wmany bins to provide for your data, but this can be modified as well
sns.histplot(x= 'bill_length_mm', data=penguins,
             bins=20)
```

```python
# If you have specific location that you'd like bins to appear, pass in list where each number are start and stop locations of the bins
# You can make bins irregularly spaced
sns.histplot(x= 'bill_length_mm', data=penguins,
             bins=[30,40,45,50,60])
```

### <code style="background:grey;color:black">binwidth</code> and <code style="background:grey;color:black">binrange</code>

```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             binwidth=3) #forces each bin to be 3 units wide
```

```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             binwidth=2, binrange=(30,60)) 
```

## Statistics

### <code style="background:grey;color:black">stat</code>

```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             stat='count') #this is the default statistic on the y axis, but this can be changed
```

#### **density** - AREA of hsitogram sums to one

##### emulate a probability density function
For each bar, if we take the height in the density scale * width of the bar, then sum all bars = 1

```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             stat='density')
```

#### **probability** - HEIGHT of histogram bars to one

##### each bar is representing the probability that a penguin would end up in a specific bucket/bin

```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             stat='probability')
```

### cumulative

##### How probability accumaluates as we move across the x axis

```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             stat='probability', fill=False, element='step',
             cumulative=True)
```

## Categorical Variables

##### If you want to split our data out by categorical variable like other seaborn plots"
- hue


```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             hue='species')
```

```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             hue='species', element='step')
```

```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             hue='species', element='poly')
```

```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             hue='species', multiple='layer') # default
```

```python
sns.histplot(x= 'bill_length_mm', data=penguins,
             hue='species', multiple='stack')
```

```python
# fill scales each bucket to range between 0-1, and shows proportion of each species in each bucket
sns.histplot(x= 'bill_length_mm', data=penguins,
             hue='species', multiple='fill')
```

### Bivariate histplot

##### We have two different variables and use colors to represent the counts

```python
sns.set_style('white')
```

```python
sns.histplot(x= 'bill_length_mm', y='bill_depth_mm', data=penguins, cbar=True)
# scale ont he right is the number of observations,s o the very dark square is 17 different penguins that have this length and depths
```

```python
sns.histplot(x= 'bill_length_mm', y='bill_depth_mm', data=penguins, hue='species')
# neat but if there's overlaps you won't be able to see the overlaps
```

```python
sns.histplot(x= 'bill_length_mm', y='species', data=penguins, hue='species', cbar=True)
```

#### **Styling**

##### **shrink**

```python
sns.histplot(x= 'species', data=penguins, hue='sex', multiple='dodge')
```

```python
# I can use shrink to space out each species on the x axis
sns.histplot(x= 'species', data=penguins, hue='sex', multiple='dodge', shrink=0.8)
```

##### **color, palette**

```python
sns.histplot(x= 'bill_length_mm', data=penguins, 
             color='indigo')
```

```python
sns.histplot(x= 'species', data=penguins, hue='sex', 
             multiple='dodge',
            shrink=0.8,
            palette='bone')
```

##### **fill**

```python
sns.histplot(x= 'bill_length_mm', data=penguins, 
             fill=False)
```
