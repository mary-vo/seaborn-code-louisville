# Seaborn: boxplot

```python
from matplotlib import pyplot as platform
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

# Basics

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

#### Filter down to only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.boxplot(cars.mpg)
```

```python
cars.mpg.describe()
```

```python
sns.boxplot(x=cars.origin, y=cars.mpg)
```

```python
# Alternatively, if data is coming from the same datafrma,e you can reference the column name and where the data is coming from
sns.boxplot(x='origin',y='mpg',data=cars)
```

# Hue

### You can separate boxplot by another category

```python
sns.boxplot(x='origin',y='mpg',hue='cylinders',data=cars)
```

### Create your own category on the fly

#### By Model Year

```python
cars.model_year.describe()
```

```python
# Create a column for newer models
cars['newer_model'] = cars.model_year > 76
```

```python
sns.boxplot(x='origin',y='mpg',hue='newer_model', data=cars);
```

# Styling Options

## Order

```python
# If you only have x and y, you can switch the order to go from verticle to horizontal bars
sns.boxplot(x='mpg',y='origin',data=cars)
```

```python
# You can use order paramter, to order bars the way you want
sns.boxplot(x='mpg',y='origin',data=cars,order=['japan', 'europe', 'usa'])
```

```python
# You can also include a hue order
sns.boxplot(x='mpg',y='origin',hue='newer_model',data=cars,order=['japan', 'europe', 'usa'],hue_order=[True,False])
```

## Color

```python
# If you want everything to be the same color
sns.boxplot(x='mpg',y='origin',data=cars,color='green')
```

```python
# If you add a hue, the hue will be a lighter color of your color setting
sns.boxplot(x='mpg',y='origin',hue='newer_model',data=cars, color='green')
```

## width

```python
sns.boxplot(x='mpg',y='origin',data=cars,width=0.2)
```

## linewidth

```python
sns.boxplot(x='mpg',y='origin',data=cars,linewidth=2.5)
```

## whis (whisker)

```python
sns.boxplot(x='mpg',y='origin',data=cars,whis=2)
```

## fliersize

### Size of the outlier diamond

```python
sns.boxplot(x='mpg',y='origin',data=cars,fliersize=10)
```

## showcaps

#### property from matplotlib

```python
sns.boxplot(x='mpg',y='origin',data=cars,showcaps=False)
```
