# Seaborn with pandas

#### Let Seaborn group, aggregate, and plot your pandas dataframes

```python
cars = sns.load_dataset('mpg')
```

```python
type(cars)
```

```python
cars.dropna(inplace=True)
cars.shape
```

```python
cars.head()
```

```python
sns.relplot(x='model_year',y='mpg',col='origin', hue='cylinders', data=cars);
```
