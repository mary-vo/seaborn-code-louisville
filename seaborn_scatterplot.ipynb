```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds_filtered = diamonds[diamonds.cut.isin(['Premium','Good']) & diamonds.color.isin(['D','F','J'])].sample(n=100,random_state=22)
diamonds_filtered.shape
```

# Basics

```python
sns.set_style('dark')
```

```python
sns.scatterplot(data=diamonds_filtered, x='carat', y='price');
```

```python
sns.scatterplot(x='x', y='y',data=diamonds_filtered)
```

# Semantic Variables

### Hue
Ability to show two different categorie

```python
sns.scatterplot(x='carat',y='price', hue='cut', hue_order = ['Premium', 'Good'], data=diamonds_filtered)
```

```python
# change color of the dots
sns.scatterplot(x='carat',y='price', hue='cut', hue_order = ['Premium', 'Good'], data=diamonds_filtered, palette=['purple','#55CCCC'])
```

```python
sns.scatterplot(x='carat',y='price', hue='depth', hue_order = ['Premium', 'Good'], data=diamonds_filtered)
```

### Style

```python
# Increase or decrease size of dot
sns.scatterplot(x='carat',y='price', data=diamonds_filtered, s=100)
```

```python
sns.scatterplot(x='carat',y='price', style='cut', style_order= ['Premium', 'Good'], data=diamonds_filtered, s=100)
```

### Size

```python
# this is the hardest for the human eye to see since the marker and color is easier to discern
sns.scatterplot(x='carat',y='price', size='cut', size_order= ['Premium', 'Good'], data=diamonds_filtered, s=100)
```

```python
sns.scatterplot(x='carat',y='price', size='cut', size_order= ['Premium', 'Good'], data=diamonds_filtered, sizes=[150,50])
```

### hue + style + size

```python
sns.scatterplot(x='carat',y='price', hue='cut', hue_order= ['Premium', 'Good'], style='cut', style_order= ['Premium', 'Good'],data=diamonds_filtered, s=100)
```

```python
sns.scatterplot(x='carat',y='price',
                hue='cut',
                hue_order= ['Premium', 'Good'],
                style='color',
                style_order= ['D', 'F','J'],
                data=diamonds_filtered,
                s=100)
```

# alpha

```python
sns.scatterplot(x='carat',y='price',
                data=diamonds_filtered,
                s=100,
                alpha=0.6) #transparency to see dots since some are overlapping
```

# marker styling

```python
# seaborn inherits from matplotlib so if you know about that package, you can also change the type of marker you are using for each dot
sns.scatterplot(x='carat',y='price',
                data=diamonds_filtered,
                s=100,
                marker='*') 
```

```python
# change edge color of the star
sns.scatterplot(x='carat',y='price',
                data=diamonds_filtered,
                s=100,
                marker='*',
                edgecolor='black') 
```
